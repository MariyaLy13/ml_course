**2. Функція з параметром**

Створіть функцію `print_square(n)`, яка виводить квадрат числа.

Наприклад: `print_square(5)` → "Квадрат числа 5 = 25"

Викличте її з різними числами.

````python
def print_square(n):
    print(f"Квадрат числа {n} = {n**2}")
list_n = [3, 5, 8, 2, 10]
for i in list_n:
    print_square(i)
````

**4. Функція пошуку максимумального значення**

Створіть функцію `find_max(numbers)`, яка:
- Приймає список чисел
- Повертає найбільше число в списку

Протестуйте на списку `[15, 42, 8, 23, 56, 19]`.


````python
def find_max(numbers):
    numbers_sorted = sorted(numbers, reverse=True)
    return numbers_sorted[0]
test_list = [15, 42, 8, 23, 56, 19]
print(find_max(test_list))
````

````python
def find_max(numbers):
    max_value = numbers[0]
    for n in numbers:
        if n > max_value:
            max_value = n
    return max_value
test_list = [15, 42, 8, 23, 56, 19]
print(find_max(test_list))
````

**5. Функція з умовами**

Створіть функцію `check_grade(grade)`, яка:
- Приймає оцінку (число)
- Повертає "Здав" якщо grade >= 60
- Повертає "Не здав" якщо grade < 60

Протестуйте на різних оцінках.


````python
def check_grade(grade: int):
    if grade >= 60 and grade <= 100:
        return "Здав"
    elif grade < 60:
        return "Не здав"
    else:
        return "Not correct input"
test_data = [45, 95, 60, 59, 110]
check_garde_results = [check_grade(x) for x in test_data]
print(check_garde_results)
````

````python
def check_grade(grade: int):
    if not 0 <= grade <= 100:
        return "Not correct input"
    return "Здав" if grade >= 60 else "Не здав"
test_data = [45, 95, 60, 59, 110]
check_garde_results = [check_grade(x) for x in test_data]
print(check_garde_results)
````
