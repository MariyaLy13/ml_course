**1. Списки - базові операції**

Створіть список температур за тиждень (у градусах Цельсія) з назвою
`temperatures` і значеннями [22, 25, 19, 23, 26, 20, 21, 22].


Виконайте (кожен пункт оцінюється в 0.5 бала):
- Виведіть перший елемент
- Виведіть останній елемент
- Додайте температуру 24
- Видаліть температуру 19
- Виведіть довжину списку
- Обчисліть суму всіх температур
- Знайдіть максимальну температуру

```python
temperatures = [22, 25, 19, 23, 26, 20, 21, 22]
```

```python
print(temperatures[0])
```

```python
print(temperatures[-1])
```

```python
temperatures.append(24)
print(temperatures)
```

```python
temperatures.remove(19)
print(temperatures)
```

```python
print(len(temperatures))
```

```python
print(sum(temperatures))
```

```python
print(max(temperatures))
```
